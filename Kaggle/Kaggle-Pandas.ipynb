```python
def count_substring(string, sub_string):
    b=0
    for i in range(len(string)-(len(sub_string)-1)):
        prueb = string[i]
        for j in range (len(sub_string)-1):
            prueb = prueb + string[i+j+1]
        if prueb == sub_string:
            b = b+1
    return b

count_substring ('abcdefcdfgcde','cde')
```

```python
s = "12:40:22AM"

def timer (s):

    x,y,z = s.split(":")

    z1 = ''.join([i for i in z if i.isdigit()])
    z2 = ''.join([i for i in z if not i.isdigit()])

    if z2 == "PM":
        s1 = str(int(x)+12)
        if s1 == "24":
            s = "12" + ":" + y + ":" + z1
            return s
        else:
            s = s1 + ":" + y + ":" + z1
            return s
    else:
        if x == "12":
            s = "00" + ":" + y + ":" + z1
            return s
        else:
            s = x + ":" + y + ":" + z1
            return s

timer(s)
```
