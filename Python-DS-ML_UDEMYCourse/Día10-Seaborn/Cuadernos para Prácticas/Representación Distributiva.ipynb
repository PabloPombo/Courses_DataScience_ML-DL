```python
import seaborn as sns
```

```python
pinguinos = sns.load_dataset("penguins")
pinguinos
```

```python
sns.displot(data=pinguinos,
           x="bill_length_mm",
           kde=True)
```

```python
sns.displot(data=pinguinos,
           x="bill_length_mm",
           kde=True,
           col="species")
```

```python
sns.displot(data=pinguinos,
           x="bill_length_mm",
           kde=True,
           col="species",
           rug=True)
```
