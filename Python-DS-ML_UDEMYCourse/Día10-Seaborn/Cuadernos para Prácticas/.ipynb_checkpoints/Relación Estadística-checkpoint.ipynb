```python
import seaborn as sns
```

```python
propinas = sns.load_dataset("tips")
propinas
```

```python
sns.relplot(data = propinas,
           x="total_bill",
           y="tip")
```

```python
sns.relplot(data = propinas,
           x="total_bill",
           y="tip",
           hue= "sex",
           size= "size")
```

```python
sns.relplot(data = propinas,
           x="total_bill",
           y="tip",
           hue= "time",
           size= "size",
           col="sex")
```
