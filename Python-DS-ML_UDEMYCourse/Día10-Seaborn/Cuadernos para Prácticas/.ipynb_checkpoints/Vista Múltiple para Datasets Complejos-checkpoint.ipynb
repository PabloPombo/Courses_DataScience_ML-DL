```python
import seaborn as sns
```

```python
pinguinos = sns.load_dataset("penguins")
```

```python
pinguinos
```

```python
sns.relplot(data=pinguinos,
           x="bill_depth_mm",
           y="bill_length_mm",
           hue="species")
```

```python
sns.jointplot(data=pinguinos,
           x="bill_depth_mm",
           y="bill_length_mm",
           hue="species")
```

```python
sns.pairplot(data=pinguinos,
            hue="species")
```
