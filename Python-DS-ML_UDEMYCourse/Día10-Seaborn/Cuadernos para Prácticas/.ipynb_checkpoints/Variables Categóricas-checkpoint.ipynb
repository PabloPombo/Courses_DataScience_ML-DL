```python
import seaborn as sns
```

```python
propinas = sns.load_dataset("tips")
```

```python
propinas
```

```python
sns.catplot(data=propinas,
            kind="box",
           x="total_bill",
            y= "sex",
           col="day")
```

```python
sns.catplot(data=propinas,
            kind="swarm",
           x="total_bill",
            y= "sex",
            hue="time",
           col="day")
```

```python
sns.catplot(data=propinas,
            kind="bar",
           x="total_bill",
            y= "sex",
           col="day")
```
