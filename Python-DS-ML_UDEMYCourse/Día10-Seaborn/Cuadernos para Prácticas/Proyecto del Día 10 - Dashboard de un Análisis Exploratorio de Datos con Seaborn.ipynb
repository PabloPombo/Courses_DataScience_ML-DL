```python
import seaborn as sns
```

```python
diamantes = sns.load_dataset("diamonds")
```

```python
diamantes
```

```python
diamantes.info()
```

```python
sns.set(style="whitegrid")
```

```python
grafica = sns.relplot(data=diamantes,
           x="carat",
           y="price",
            height=6,
            aspect=1.5)

grafica.set_xlabels("Peso en quilates")
grafica.set_ylabels("Precio")

grafica.fig.suptitle("Relacion entre Pecio y Peso en quilates",
                    ha='center',
                    va='baseline')
```

```python
grafica2 = sns.displot(data=diamantes,
           x="price",
           kde=True,
            height=6,
            aspect=1.5)
grafica2.fig.suptitle("Frecuencia Precios")
grafica2.set_axis_labels("Precio", "Frecuencia")
```

```python
grafica3 = sns.histplot(data=diamantes,
                       x="carat",
                       kde=True,
                       color="green")
grafica3.set_title("Frecuencia Quilates")
grafica3.set(xlabel='Quilates', ylabel='Frecuencia');
```

```python
grafica4 = sns.catplot(data=diamantes,
           y="price",
            x="cut",
           kind='box')


grafica4.fig.suptitle('Precio de los Diamantes por Calidad del Corte')

grafica4.set_axis_labels("Calidad del Corte",
                  "Precio");
```

```python
grafica5 = sns.jointplot(data=diamantes,
                  x="depth",
                  y="price",
                 kind="kde")
grafica5.fig.suptitle("Relacion entre profundidar y precio")

grafica5.set_axis_labels("Profundidad", "Precio")
```

```python
sns.pairplot(data=diamantes, corner=True)
```
