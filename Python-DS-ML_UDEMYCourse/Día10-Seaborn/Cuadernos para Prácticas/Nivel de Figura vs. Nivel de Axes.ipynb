```python
import seaborn as sns
```

```python
pinguinos = sns.load_dataset("penguins")
```

```python
pinguinos
```

```python
plot1 = sns.displot(data=pinguinos,
           x="flipper_length_mm",
           hue="species",
           multiple="stack")
```

```python
plot2= sns.relplot(data=pinguinos,
           x="flipper_length_mm",
            y="bill_depth_mm",
           hue="species")
```

```python
plot1 = sns.displot(data=pinguinos,
           x="flipper_length_mm",
           hue="species",
           multiple="stack")
plot1.set_axis_labels("Flipper length (mm)", "Conteo")
```

```python
plot2= sns.relplot(data=pinguinos,
           x="flipper_length_mm",
            y="bill_depth_mm",
           hue="species")
plot2.legend.set_title("Especies")
```
