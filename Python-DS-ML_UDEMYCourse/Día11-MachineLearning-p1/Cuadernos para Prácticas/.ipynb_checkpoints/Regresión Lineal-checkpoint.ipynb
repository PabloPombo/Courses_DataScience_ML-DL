```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.DataFrame()
df["Area"] = [2600,3000,3200,3600,4000]
df["Precio"] = [550000,565000,610000,680000,725000]

X = df[['Area']]
y = df['Precio']

model = linear_model.LinearRegression()
model.fit(X,y)

model.predict([[300]])
```

```python
plt.plot(df['Area'],model.predict(X))
plt.title('Precio unidades')
plt.grid()
plt.xlabel('Area')
plt.ylabel('Precio')

plt.scatter(df['Area'],df['Precio'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import linear_model

# Crear DataFrame
data = {'Edad': [25, 30, 35, 40, 45],
        'Salario': [3000, 3500, 4000, 4500, 5000]}
df = pd.DataFrame(data)

X = df[['Edad']]
y = df['Salario']

modelo = linear_model.LinearRegression()
modelo.fit(X,y)

# Crear el grfico
plt.title('Relacin entre Edad y Salario')
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.grid(True)
plt.plot(df["Edad"],
        modelo.predict(X));
        
plt.scatter(df['Edad'], df['Salario'], color='blue', marker='o')
```
