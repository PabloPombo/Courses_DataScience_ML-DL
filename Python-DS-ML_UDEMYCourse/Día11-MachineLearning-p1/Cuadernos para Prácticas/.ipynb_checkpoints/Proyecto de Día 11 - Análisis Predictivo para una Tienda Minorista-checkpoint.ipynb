```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree, DecisionTreeRegressor
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split

ruta = "C:/Users/Pablo/Documents/CURSOS-Aprendizaje/Courses_DataScience_ML-DL/Python-DS-ML_UDEMYCourse/Día11-MachineLearning-p1/Ventas.csv"
df = pd.read_csv(ruta)
```

```python
df.info()
```

```python
df.head(5)
```

```python
df["Fecha"] = pd.to_datetime(df["Fecha"], format='%Y/%m/%d')
df
```

```python
columnas_escalar = df.drop(["Fecha","Ventas"], axis=1).columns

escala = MinMaxScaler(feature_range=(0, 1))
normado = escala.fit_transform(df[columnas_escalar])
normado
```

```python
df_normado = pd.DataFrame(normado, columns=columnas_escalar)
df_normado
```

```python
X = df_normado
y = df["Ventas"]
```

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X,y, train_size=0.8, random_state=42)
```

```python
cols = X.columns
cols
```

```python
modelos = [
    ("Regresion Lineal", LinearRegression()),
    ("Arbol de Decision", DecisionTreeRegressor()),
    ("Random Forest", RandomForestRegressor())
]

modelos
```

```python
for nombre, modelo in modelos:
    modelo.fit(X_entrena, y_entrena)
    califica = modelo.score(X_prueba,y_prueba)
    print(f'{nombre}: {califica}')
```

```python
mod = LinearRegression()
mod.fit(X_entrena,y_entrena)
```

```python
prediccion = mod.predict(X_prueba)
prediccion
```

```python
plt.figure(figsize=(10,10))
plt.scatter(y_prueba,prediccion, alpha=0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()])
plt.show()
```
