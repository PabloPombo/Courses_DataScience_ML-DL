```python
df["Fecha"] = pd.to_datetime(df["Fecha"], format='%Y/%m/%d')
df
```

```python
columnas_escalar = df.drop(["Fecha","Ventas"], axis=1).columns

escala = MinMaxScaler(feature_range=(0, 1))
normado = escala.fit_transform(df[columnas_escalar])
normado
```

```python
df_normado = pd.DataFrame(normado, columns=columnas_escalar)
df_normado
```

```python
X = df_normado
y = df["Ventas"]
```

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X,y, train_size=0.8, random_state=42)
```

```python
cols = X.columns
cols
```

```python
for nombre, modelo in modelos:
    modelo.fit(X_entrena, y_entrena)
    califica = modelo.score(X_prueba,y_prueba)
    print(f'{nombre}: {califica}')
```
