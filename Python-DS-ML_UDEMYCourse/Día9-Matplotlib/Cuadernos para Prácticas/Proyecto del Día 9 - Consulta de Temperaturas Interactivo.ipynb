```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Pablo/Documents/CURSOS/Cursos UDEMY/Curso_Python-DS-ML/Curso_Python-DS-ML/Día9-Matplotlib/Datos+Meteorológicos_Arg_2023.csv") 
df
```

```python
df ['Fecha'] = pd.to_datetime(df['Fecha'], format='%d/%m/%Y')
```

```python
df.info()
```

```python

def programa():
    a = 0 
    meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Nobiembre','Diciembre']
    ciudades = df['Ciudad'].unique()
    
    while True:
        
        print(ciudades)
        ciudad = input('Escribe el nombre de una Ciudad para investigar las temperaturas: ')
        mes = int(input('Escribe el numero del mes del que quieras saber: '))
        
        datos_ciudad_mes = df[(df['Ciudad'] == ciudad) & (df['Fecha'].dt.month == mes)]
    
        plt.figure(figsize=(10, 6))
        plt.plot(datos_ciudad_mes['Fecha'],datos_ciudad_mes['Temperatura Maxima'], label='Temperatura Maxima')
        plt.plot(datos_ciudad_mes['Fecha'],datos_ciudad_mes['Temperatura Minima'], label='Temperatura Minima')
        plt.title(f'Grafica de la ciudad {ciudad} del mes de {meses[int(mes)-1]}')
        plt.xlabel('Fecha')
        plt.ylabel('Temperatura')
        plt.xticks(rotation=45)
        plt.legend()
        plt.show()
        
        continuar = input('Si deseas realizar otra consulta teclea "y", sino teclea cualquier otra letra: ')
        if continuar !='y':
            break
```

```python
programa()
```
