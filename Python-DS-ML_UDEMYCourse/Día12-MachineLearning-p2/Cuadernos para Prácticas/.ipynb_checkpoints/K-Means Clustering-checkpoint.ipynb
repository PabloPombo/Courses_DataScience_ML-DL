```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
np.random.seed(42)
numeros = np.random.rand(100,2)
numeros
```

```python
modelo = KMeans(n_clusters=5, random_state=42)
modelo.fit(numeros)
```

```python
etiquetas = modelo.labels_
centroides = modelo.cluster_centers_
centroides
```

```python
etiquetas
```

```python
plt.figure(figsize=(10,6))
plt.scatter(numeros[:,0],numeros[:,1], marker='o', c=etiquetas)
plt.scatter(centroides[:,0],centroides[:,1], marker='x', s=200, color='red', label='centroides')
plt.legend()
plt.show()
```

```python
pinguinos = sns.load_dataset("penguins")
pinguinos.head()
```

```python
pinguinos = pinguinos.dropna()
```

```python
data = pinguinos[["bill_length_mm","bill_depth_mm"]]
data
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data)
```

```python
centroides = kmeans.cluster_centers_
etiquetas = kmeans.labels_
centroides
```

```python
etiquetas
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data["bill_length_mm"], data["bill_depth_mm"], c=etiquetas, marker='o')
plt.scatter(centroides[:,0], centroides[:,1], c='red', s=200, marker='X', label="Centroides")
plt.legend()
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(data=data, x="bill_length_mm", y="bill_depth_mm", hue=etiquetas)
plt.scatter(centroides[:,0], centroides[:,1], c='red', s=100, marker='X', label="Centroides")
plt.legend()
```

```python
import numpy as np 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

np.random.seed(42)

datos = np.random.rand(100,2)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(datos)

etiquetas = kmeans.labels_
centroides = kmeans.cluster_centers_

plt.figure(figsize=(10,6))
plt.scatter(datos[:,0], datos[:,1], c=etiquetas,  marker='o', cmap='viridis')
plt.scatter(centroides[:,0], centroides[:,1], c='red', s=200, marker='x', label='Centroides')
plt.title('K-Means Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```
